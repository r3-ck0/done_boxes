```python
import angr, claripy
from pwn import *
```

```python
elf = ELF("./gates")
```

```python
!objdump -d ./gates
```

```python
p = angr.Project("./gates", main_opts={"base_addr": 0x100000}, auto_load_libs=False)
```

```python
s = p.factory.entry_state(addr=0x1010bc)
sm = p.factory.simgr(s)

vs = [claripy.BVS(f"chr{x}", 8) for x in range(32)]

def prepare(state):
    for i, v in enumerate(vs):
        state.memory.store(0x10404c + (i * 0x10), 1, 1)
        state.memory.store(0x10404c + 1 + (i * 0x10), v, 1)

p.hook(0x1010ba, prepare, length=0x1010bc-0x1010ba)
```

```python
results = sm.explore(find=lambda s: b"Correct" in s.posix.dumps(1), avoid=lambda s: b"Wrong" in s.posix.dumps(1))
```
