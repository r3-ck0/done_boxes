```python
p = angr.Project("./p2", main_opts={"base_addr": 0}, auto_load_libs=False)

user_ipt = claripy.BVS("user_ipt", 10*8)

s = p.factory.entry_state(addr=elf.symbols["main"], stdin=user_ipt)
sm = p.factory.simgr(s)

class nothing(angr.SimProcedure):
    def run(self):
        pass
        
class read(angr.SimProcedure):
    def run(self):
        self.state.memory.store(self.state.regs.rdx, self.state.memory.load(0x20000000, 4), 4)

class write(angr.SimProcedure):
    def run(self):
        self.state.memory.store(0x20000000, self.state.regs.rdx, 4)

p.hook_symbol('fprintf', write())
p.hook_symbol('fclose', nothing())
p.hook_symbol('fopen', nothing())
p.hook_symbol('__isoc99_fscanf', read())
```

```python
sm.explore(find=0x12b0, avoid=0x12c1)
```
